```python
import tft.ql.expr as ql
from tft.ql.util import *
from tft.client.meta import *
```

```python
client = MetaTFTClient()
comps = client.fetch(MetaTFTApis.COMPS_DATA)
set_data = client.fetch(MetaTFTApis.SET_DATA)
```

```python
q = ql.query(comps)
q2 = ql.query(set_data)
```

```python
# All item data.
ql.trav('items').eval(set_data)
```

```python
# Buildable items.
buildable_items = q2.trav('items').filter(ql.trav('composition').len().eq(2))
buildable_items.map(ql.trav('apiName')).vals().eval() # Names.
```

```python
# Base components names.
component_names = buildable_items.map(ql.trav('composition')).vals().flatten().uniq().eval()
component_items = q2.trav('items').filter(ql.trav('apiName').in_set(component_names))
component_items.map(ql.trav('apiName')).vals().eval()
```

```python
# Emblem names.
q2.trav('items')
```

```python
# TFT set
q.trav('tft_set').eval()
```

```python
q.trav('results.data.cluster_details.0.units_string').splay(depth=0)
```
